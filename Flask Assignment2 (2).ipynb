### Q1. Explain GET and POST methods.

GET and POST are methods used to obtain data.The difference between them is that when we pass data through get the data is visible in URl and it is obtained through URL while in POST method the data is not visible in the URL and it is through body.
EXAMPLES: GET- Google Search, POST- login id and password

GET is a request method used to retrieve or fetch data from a web server.
It is a safe and idempotent method, which means it should not modify the state of the server or the data on the server.
GET requests can be cached by browsers, and the URLs are visible in the browser's address bar.

POST is a request method used to submit data to a web server to be processed and stored.
It is not safe or idempotent, as it may modify the state of the server or the data on the server.
POST requests are not cached by browsers, and the data is not visible in the browser's address bar.

### Q2. Why is request used in Flask?

Accessing request data: The request object provides access to various data associated with the incoming HTTP request, such as query parameters, form data, cookies, headers, and files. This data can be accessed from the request object in Flask, allowing the application to extract and process the data as needed.
 
Handling different HTTP methods: The request object allows Flask applications to handle different HTTP methods, such as GET, POST, PUT, DELETE, etc. The request.method attribute in Flask provides the current HTTP method used in the request, allowing the application to conditionally handle different methods and perform corresponding actions.

### Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the user's browser to a different URL or route. It is commonly used in web applications for various purposes

When a user accesses a particular URL or route in a Flask application, the redirect() function can be used to redirect the user to a different URL or route. This can be useful for redirecting users after successful login, logout, or form submission, or for handling URL redirects for SEO purposes.

 The redirect() function can be used to handle errors and exceptions in a Flask application. For example, when an error occurs, such as a page not found (404) error, the redirect() function can be used to redirect the user to a custom error page or a designated error handling route.

 The redirect() function can be used in Flask applications for user authentication and authorization. For example, if a user tries to access a protected resource without proper authentication or authorization, the redirect() function can be used to redirect them to a login page or a designated access denied page.

### Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used for rendering dynamic HTML pages. Templates are essentially HTML files that contain placeholders or variables, which are dynamically replaced with actual values when the template is rendered. Flask uses a template engine, such as Jinja2, to render templates and generate HTML responses to be sent to the client's browser.

The render_template() function in Flask is used to render templates and generate HTML responses. It takes a template file as input, along with any required data or variables, and processes the template to generate the final HTML output. The rendered HTML is then sent as a response to the client's browser.
